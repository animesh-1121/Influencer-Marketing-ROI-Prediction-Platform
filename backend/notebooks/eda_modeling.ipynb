# Phase 1 & 2: EDA and Model Building
This notebook handles data loading, exploratory data analysis (EDA), feature engineering, and training models.

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/influencer_dataset.csv')
if 'Likes' not in df.columns:
    np.random.seed(42)
    df['Likes'] = (df['Engagement'] * np.random.uniform(0.7, 0.9, len(df))).astype(int)
    df['Comments'] = df['Engagement'] - df['Likes']

df.head()
```

```python
plt.figure()
sns.histplot(df['Followers'], bins=30, kde=True)
plt.title('Followers Distribution')
plt.show()
```
